````python
# ğŸ—‚ï¸ 3. Dataset Setup
projector_aircon/
â”œâ”€â”€ images-train/
â”œâ”€â”€ images-val/
â”œâ”€â”€ labels-train/
â”œâ”€â”€ labels-val/
â””â”€â”€ data.yaml
````

````python
# ğŸ‹ï¸ 4. Train the Model
# Load a lightweight YOLOv8 model

model = YOLO("yolov8n.pt")

# Train on custom dataset
model.train(
    data="/content/projector_aircon/data.yaml",
    epochs=30,
    imgsz=320,
    batch=16,
    device=0   # GPU (set 'cpu' if no GPU)
)
````

````python
#ğŸ” 5. Run Inference
## runs/detect/train/weights/best.pt

# Load trained model
model = YOLO("runs/detect/train/weights/best.pt")

# Run inference on validation images
results = model.predict(source="/content/projector_aircon/images-val")
````

````python
#ğŸ“Š 6. Visualize Results
# Display results
for r in results:
    r.show()  # show detections on images
````
